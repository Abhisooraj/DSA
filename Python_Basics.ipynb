## **LIST**

```python
a=[10,20,30,"Abhishek",100.25,3+3j]
print(a)
print(a[3])
print(a[3][1])
print(a[1:4])
print(a[-2])
b=list("Abhishek")
print(b)
```

```python
del a[2]
print(a)
```

```python
c=[50, 60,70]
c.append(a)
print(c)
print(a)
```

```python
d=[x for x in c[3]]
print(d[2])
# c[3] is the list a, and d becomes its copy.
```

```python
c.insert(3,"Sooraj") # insert(3, ...) inserts "Sooraj" at index 3, shifting the existing element to the right.
print(c)
```

```python
c.pop() # pop() removes and returns the last element of the list, which is the nested list a.
# c becomes: [50, 60, 70, 'Sooraj']
```

```python
c.reverse()
print(c)
```

```python
x=["DSA","Python","Programming"]
c.extend(x) # After extend(x), each element of x is added individually to the end of c.
print(c)
```

# Python List Functions

- append()   → Add a single element
- extend()   → Add multiple elements
- insert()   → Insert an element at a specific index
- remove()   → Remove the first occurrence of an element
- pop()      → Remove and return an element
- clear()    → Remove all elements from the list
- index()    → Return the index of an element
- count()    → Count occurrences of an element
- sort()     → Sort the list
- reverse()  → Reverse the list
- copy()     → Create a shallow copy of the list

## **List Comprehension**

```python
a=[x for x in range(10)]
print(a)
```

```python
b=[x**x for x in range(5)]
print(b)
```

```python
k=[p for p in a if p%2==0]
print(k)
```

```python
l=[i for i in range(2,8) for i in range(i*2,50,i)]
print(l)
j=[m for m in range(2,50)if m not in l]
print("Prime numbers: ",j)
```
