## **Two Sum**
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target. <br>You may assume that each input would have exactly one solution, and you may not use the same element twice.

```python
def twoSum(nums, target):
    for i in range(len(nums)):
        for j in range(i):
            if nums[i] + nums[j] == target:
                return [i, j]

    return []


# Driver Code
nums = list(map(int, input("Enter numbers: ").split()))
target = int(input("Enter target: "))

result = twoSum(nums, target)

print("Indices:", result)
```

```python
# Optimal Solution
def twoSum(nums, target):
    mp = {}

    for i in range(len(nums)):
        diff = target - nums[i]

        if diff in mp:
            return [mp[diff], i]

        mp[nums[i]] = i

    return []

# Driver Code
nums = list(map(int, input("Enter numbers separated by space: ").split()))
target = int(input("Enter target: "))

result = twoSum(nums, target)

print("Indices:", result)
```

## Check Palindrom

```python
# Naive solution
class Solution:
    def isPalindrome(self, n: int) -> bool:
        if n < 0:
            return False

        original = n
        rev = 0

        while n > 0:
            rem = n % 10
            rev = rev * 10 + rem
            n = n // 10

        if rev == original:
            return True
        else:
            return False


# Driver code
n = int(input("Enter Number: "))
obj = Solution()
print(obj.isPalindrome(n))
```

```python
# optimal solution
class Solution:
    def isPalindrome(self, n: int) -> bool:
        if n < 0:
            return False
        # Numbers ending with 0 (except 0 itself) are not palindromes
        if n%10==0 and n!=0:
          return False
        rev=0
        while n > rev:
            rem = n % 10
            rev = rev * 10 + rem
            n = n // 10

        return n == rev or n == rev // 10

# Driver code
n = int(input("Enter Number: "))
obj = Solution()
print(obj.isPalindrome(n))
```
