## Level 1: Conversion Practice

Print 1 to N

```python
n=int(input("Enter n: "))
for i in range(1,n+1):
  print(i,end=" ")
```

Print reverse (N to 1)

```python
n=int(input("Enter n: "))
for i in range(n,0,-1):
  print(i,end=" ")
```

Sum of N numbers

```python
n=int(input("Enter n: "))
total=0
for i in range(1,n+1):
  total+=i
print(total,end=" ")
```

Factorial

```python
n=int(input("Enter n: "))
fact=1
for i in range(1,n+1):
  fact*=i
print(fact,end=" ")
```

## Level 2: Smart Iteration



Print all divisors of a number

```python
n=int(input("Enter a number: "))
for i in range(1,n+1):
  if n%i==0:
    print(i,end=" ")
```

Count Divisors

```python
n=int(input("Enter a number: "))
count=0
for i in range(1,n+1):
  if n%i==0:
    count+=1
print("Count divisors: ",count)
```

Check Prime (using √n optimization)

```python
from math import sqrt
num=int(input("Enter number: "))
if num<=1:
    print("Number is not prime")
else:
  for i in range(2, int(sqrt(num))+1):
    if num%i==0:
      print("Number is not prime")
      break
  else:
    print("Number is prime")
```

Find GCD using loop

```python
num1= int(input("Enter number 1: "))
num2= int(input("Enter number 2: "))
def GCD(num1,num2):
  if (num1==0 or num2==0):
    return 0
  if num1 == num2:
    return num1
  if num1>num2:
    return GCD(num1-num2,num2)
  return GCD(num1,num2-num1)
print("GCD =", GCD(num1, num2))
```
