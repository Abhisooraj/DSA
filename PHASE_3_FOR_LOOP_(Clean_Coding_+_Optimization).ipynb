## Level 1: Conversion Practice

Print 1 to N

```python
n=int(input("Enter n: "))
for i in range(1,n+1):
  print(i,end=" ")
```

Print reverse (N to 1)

```python
n=int(input("Enter n: "))
for i in range(n,0,-1):
  print(i,end=" ")
```

Sum of N numbers

```python
n=int(input("Enter n: "))
total=0
for i in range(1,n+1):
  total+=i
print(total,end=" ")
```

Factorial

```python
n=int(input("Enter n: "))
fact=1
for i in range(1,n+1):
  fact*=i
print(fact,end=" ")
```

Print all divisors of a number

```python
n=int(input("Enter a number: "))
for i in range(1,n+1):
  if n%i==0:
    print(i,end=" ")
```

Count Divisors

```python
n=int(input("Enter a number: "))
count=0
for i in range(1,n+1):
  if n%i==0:
    count+=1
print("Count divisors: ",count)
```
