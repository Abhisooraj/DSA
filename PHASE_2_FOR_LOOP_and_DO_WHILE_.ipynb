# PHASE 2: DO-WHILE + CONTROL FLOW

## Level 1: Basic do-while Thinking

Print numbers from 1 to N (execute at least once)

```python
i=1
n= int(input("Enter N: "))
while(True):
  print(i,end=" ")
  i+=1
  if(i>=n):
    break
```

Take input until user enters 0

```python
while(True):
  n=int(input("Enter value: "))
  if n==0:
    break
```

Print table of a number (at least once)

```python
i=1
n=9
while(i<=10):
  print(n*i)
  i+=1
```

##  Level 2: Input Control

Keep asking user for number until they enter positive number

```python
while(True):
  n=int(input("Enter value: "))
  if n>0:
    break
```

Menu-driven program:

1 → Add
2 → Subtract
3 → Exit

```python
while (True):
  n=int(input("Menu-driven program:\n1 → Add\n2 → Subtract\n3 → Exit: "))
  if n==3 or n>3:
    break
  n1=int(input("Enter value 1: "))
  n2=int(input("Enter value 2: "))
  if n==1:
    print(n1+n2)
    continue
  if n==2:
    print(n1-n2)
    continue
```

## Level 3: Validation Problems

Keep taking numbers and print their sum until user enters -1

```python
total=0
print("Enter -1 to get out of the loop")
while True:
  n=int(input("Enter number: "))
  if n==-1:
    break
  total+=n
  print("Sum: ",total)
```

Count how many inputs user entered before stopping

```python
count=0
print("Enter -1 to get out of the loop")
while True:
  n=int(input("Enter number: "))
  count+=1
  if n==-1:
    break
print("count: ",count-1)
```

Find maximum number entered (until user stops)

```python
maximum=None
print("Enter -1 to get out of the loop")
while True:
  n=int(input("Enter number: "))
  if n==-1:
    break
  if maximum is None or n>maximum:
    maximum=n
print("maximum: ",maximum)
```

## Level 4: Mixed Logic

Guess the number (loop until correct)

```python
import random

secret = random.randint(1, 10)

while True:
    guess = int(input("Guess the number (1–10): "))

    if guess == secret:
        print("Correct! You guessed it ")
        break
    elif guess < secret:
        print("Too low, try again.")
    else:
        print("Too high, try again.")
```

Keep asking password until correct

```python
password="python"
while True:
    guess = input("Guess the password: ")

    if guess == password:
        print("Correct! You guessed it ")
        break
```

# PHASE 3: FOR LOOP (Clean Coding + Optimization)

## Level 1: Conversion Practice

Print 1 to N


```python
n=int(input("Enter n: "))
for i in range(1,n+1):
  print(i,end=" ")
```

Print reverse (N to 1)

```python
n=int(input("Enter n: "))
for i in range(n,0,-1):
  print(i,end=" ")
```

Sum of N numbers

```python
n=int(input("Enter n: "))
total=0
for i in range(1,n+1):
  total+=i
print(total,end=" ")
```

Factorial

```python
n=int(input("Enter n: "))
fact=1
for i in range(1,n+1):
  fact*=i
print(fact,end=" ")
```
