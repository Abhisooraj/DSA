```python
num=int(input("Enter Number: "))
rev=0
n=num
while(n>0):
  rem=n%10
  rev=rev*10+rem
  n=n//10
if (rev==num):
  print("PALINDROME")
else:
  print("not PALINDROME")
```

Check Armstrong number

```python
def countDigits(n):
  if n == 0:
        return 1
  count=0
  while n>0:
    n = n//10
    count+=1
  return count

def checkArmstrong(n):
  digits=countDigits(n)
  temp=n
  sum=0
  while temp>0:
    rem=temp%10
    sum+=rem**digits
    temp=temp//10

  if sum==n:
     return sum
```

```python
i=1
n=int(input("ENter range from 1 to n: "))
while(i<=n):
  if checkArmstrong(i):
    print(i,end=" ")
  i+=1
```

Check Perfect number

```python
n=int(input("Enter a Number: "))
i=1
total=0
while i<n:
  if n%i==0:
    total+=i
  i+=1
if total==n:
  print("perfect number")
else:
  print("Not a perfect number")
```

# Level 5: Series

Print Fibonacci series up to N terms

```python
def fibonacciSeries(n):
  a=0
  b=1
  i=0
  while i<n:
    a,b=b,a+b
    print(a,end=" ")
    i+=1

fibonacciSeries(10)
```

# Level 6: Power & Factorial

Find factorial of a number

```python
def factorial(n):
  i=1
  fact=1
  while i<=n:
    fact *=i
    i+=1
  print(fact)
factorial(5)
```

Calculate a^b using loop

```python
a=int(input("Enter a: "))
b=int(input("Enter b: "))
i=0
res=1
while i<b:
  res*=a
  i+=1
print(res)
```

Print squares from 1 to N

```python
def printSquare(n):
  i=1
  if n<0:
    print("Not possible")
  else:
    while i<=n:
      print(i*i,end=" ")
      i+=1

printSquare(5)
```

Print cubes from 1 to N

```python
def printCube(n):
  i=1
  if n<0:
    print("Not possible")
  else:
    while i<=n:
      print(i*i*i,end=" ")
      i+=1

printCube(5)
```

# Level 7: Mixed Logic

Print numbers between A and B

```python
n1=int(input("Enter range 1: "))
n2=int(input("Enter range 2: "))
if n1 <= n2:
  while(n1<=n2):
    print(n1,end=" ")
    n1+=1
else:
  while(n1>=n2):
    print(n1,end=" ")
    n1-=1
```

```python
count=0
n1=int(input("Enter range 1: "))
n2=int(input("Enter range 2: "))
while(n1<=n2):
  if ((n1%2==0 and n1%3==0)):
    print(n1, end=" ")
    count+=1
  n1+=1
print("\nCount:", count)
```

Find largest digit in a number

```python
n=int(input("Enter number: "))
largest=0
while n>0:
  rem=n%10
  if rem>largest:
    largest=rem
  n=n//10
print("largest: ",largest,end=" ")
```

Find smallest digit in a number

```python
n=int(input("Enter number: "))
smallest=n%10 # Initialize with last digit:
while n>0:
  rem=n%10
  if rem<smallest:
    smallest=rem
  n=n//10
print("smallest: ",smallest,end=" ")
```
